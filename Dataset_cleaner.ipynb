```python
import pandas as pd

file_path = 'D:/Super_weekend/Super_weekend/Datasets/dataset.csv'
data = pd.read_csv(file_path)
```

```python
# Display the first few rows and general information about the dataset
data.head(), data.info()
```

```python
date_columns = [
    "SHIFT_START_AT", "SHIFT_CREATED_AT", "OFFER_VIEWED_AT", 
    "CLAIMED_AT", "CANCELED_AT", "DELETED_AT"
]
for col in date_columns:
    data[col] = pd.to_datetime(data[col], errors='coerce')
```

```python
data.head()
```

```python
data["DURATION"] = data["DURATION"].fillna(0)
```

```python
data.head()
```

```python
data["IS_CLAIMED"] = data["CLAIMED_AT"].notnull()
```

```python
data["POST_TO_VIEW_TIME"] = round((data["OFFER_VIEWED_AT"] - data["SHIFT_CREATED_AT"]).dt.total_seconds() / 3600, 2)
```

```python
data["VIEW_TO_CLAIM_TIME"] = round((data["CLAIMED_AT"] - data["OFFER_VIEWED_AT"]).dt.total_seconds() / 3600, 2)
```

```python
data.head()
```

```python
data.info()
```

```python
data = data[data["DURATION"] > 0]
```

```python
data.info()
```

```python
data = data.drop_duplicates()
```

```python
data.info()
```
